```python
import covariance_learn
import itertools
from mpl_toolkits.mplot3d import Axes3D
import phase_transition
import htree
import matplotlib.pyplot as plt
import numpy as np
import scipy.linalg
%matplotlib inline
# from IPython.html.widgets import interact
```

```python
z1 = 2
z2 = 2
tree = htree.construct_tree(arity=4, depth=2)
# Create sample

a = -.4
b = .28
Theta = phase_transition._get_mx(a, b, mx_type='smith')

n_samples = 16

X = np.random.normal(size=(n_samples, Theta.shape[0]))

eigvals, eigvecs = scipy.linalg.eigh(Theta)
print np.min(eigvals)
C = eigvecs.dot(np.diag(1 / np.sqrt(eigvals))).dot(eigvecs.T)

# X = X - X.mean(axis=0)
X = X.dot(C)

Y = C

def cov2corr(cov):
    return np.diag(1 / np.sqrt(np.diag(cov))).dot(cov.dot(np.diag(1 / np.sqrt(np.diag(cov)))))

plt.matshow(Theta, cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('population precision matrix')
plt.matshow(np.linalg.inv(cov2corr(X.T.dot(X))), cmap=plt.cm.RdBu_r)
plt.clim((-2, 2))
plt.colorbar()
plt.title('sample (scaled) precision matrix')
plt.matshow(cov2corr(Y.T.dot(Y)), cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('population correlation matrix')
plt.matshow(cov2corr(X.T.dot(X)), cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('sample correlation matrix')
```

## Estimator of the precision matrix
The estimator of $\Theta$ is based on a hierarchical graphical lasso with embedded hierarchies.
The penalty function is $$\lambda^{\frac{\text{lev}_{\text{max}} - 1 + 10^h}{\text{lev} - 1 + 10^h}}$$ where $h$ is a free parameter, $\lambda$ is the usual penalisation, and lev is the level (depth) in the tree.

```python
def estimate_precision(alpha, h, method):
    max_level = 2
    alpha_func = lambda a, lev: a ** ((max_level - 1 + 10 ** h) / (lev - 1 + 10 ** h))
    if method == 'gl':
        covl = covariance_learn.GraphLasso(alpha=alpha, score_norm='KL', max_iter = 1e4, rho=2, verbose=50)
    elif method == 'hgl':
        covl = covariance_learn.HierarchicalGraphLasso(alpha=alpha, htree=tree, alpha_func=alpha_func, score_norm='KL',
                                                       max_iter = 1e4, rho=2, verbose=50)
    covl.fit(X)
    K = covl.precision_
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.matshow(np.ma.masked_equal(K * (covl.auxiliary_prec_ != 0), 0), cmap=plt.cm.RdBu_r, vmin=-np.max(np.abs(K)) / 5.,
                vmax=np.max(np.abs(K)) / 5.)
    # ax1.set_clim((-np.max(np.abs(K)), np.max(np.abs(K))))  # (-1, 1))  # 
    # ax1.colorbar()
    ax1.set_title('Estimated precision')
    ax1.set_axis_bgcolor('.7')
    # inspire colormap by http://scikit-learn.org/stable/auto_examples/covariance/plot_sparse_cov.html
    ax2.matshow(np.array(covl.auxiliary_prec_ != 0, dtype=float), cmap=plt.cm.gray, vmin=0, vmax=1)
    DiagK = np.diag(np.diag(K))
    ax2.set_title('Support (derived from split variable)')
    fig, ax = plt.subplots()
    plt.plot(covl.f_vals_, 'b-')
    plt.title("Jensen's divergence = {}".format(covl.score(Y)))
    ax.set_ylabel('energy', color='b')
    for tl in ax.get_yticklabels():
        tl.set_color('b')
    ax2 = ax.twinx()
    ax2.semilogy(covl.var_gap_, 'g-')
    ax2.semilogy(covl.dual_gap_, 'g--')
    ax2.set_ylabel('split diff & dual update norm', color='g')
    for tl in ax2.get_yticklabels():
        tl.set_color('g')

# interact(estimate_precision, alpha=(0, 1., .01), h=(-2, 2, 1. / 8), method=('hgl', 'gl'))
```

Hierarchical Graph Lasso with $\lambda=1/1000$ and $h=6/10$, resulting in a suppression of some off-diagonal blocks that are supposed non-active

```python
estimate_precision(alpha=.06, h=1, method='hgl')
```

Hierarchical Graph Lasso with $\lambda=1/1000$ and $h=-100$, which approaches the Graphical Lasso for a tree with depth two.

```python
estimate_precision(alpha=.2, h=1, method='hgl')
```

Indeed, the Graphical Lasso with the same $\lambda$ as in the above example of Hierarchical Graph Lasso yields the same results when the latter has its parameter $h\to -\infty$

```python
estimate_precision(alpha=.2, h=-1, method='gl')
```

To suppress the blocks in the Graphical Lasso we need to choose a higher $\lambda$ penalising the diagonal blocks more. It is not sure that a such $\lambda\neq\lambda_{\text{max}}$ actually exists.

```python
estimate_precision(alpha=.06, h=-10, method='gl')
```

```python
def alpha_func_(h, max_level):
    return lambda a, lev: min(1., a ** ((max_level - 1 + 10 ** h) / (lev - 1 + 10 ** h)))

max_level = max([lev for (_, lev) in tree.root_.get_descendants()])
# print "tree with {} levels".format(max_level)

n_h = 11
n_a = 11
h_vals = np.linspace(-1, 1, n_h)
h_ix = np.arange(n_h)
alpha_vals = np.logspace(-2, 0, n_a)
a_ix = np.arange(n_a)
scores = np.zeros((n_h, n_a))
score_gl = np.zeros((n_h, n_a))

for (ix, vals) in zip(itertools.product(h_ix, a_ix), itertools.product(h_vals, alpha_vals)):
    # print "computing h={}, alpha={} yielding ({}, {})".format(vals[0], vals[1],
    #                                                          alpha_func_(vals[0], max_level)(vals[1], 2),
    #                                                          alpha_func_(vals[0], max_level)(vals[1], 1))
    if ix[0] == 0:
        covl = covariance_learn.GraphLasso(
            alpha=vals[1], score_norm='KL', max_iter=1e4, rho=2, verbose=50)
        score_gl[..., ix[1]] = covl.fit(X).score(Y)
    covl = covariance_learn.HierarchicalGraphLasso(
        alpha=vals[1], htree=tree, alpha_func=alpha_func_(vals[0], max_level),
        score_norm='KL', max_iter=1e4, rho=2, verbose=50)
    scores[ix] = covl.fit(X).score(Y)
    # print "\tscore: {}".format(scores[ix])
```

```python
fig = plt.figure()
plot_scores = np.log(scores / score_gl).T
ax = fig.gca(projection='3d')
X_, Y_ = np.meshgrid(h_vals, alpha_vals)
surf = ax.plot_surface(X_, Y_, plot_scores, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cset = ax.contour(X_, Y_, plot_scores, zdir='z', offset=-1, cmap=plt.cm.coolwarm)
cset = ax.contour(X_, Y_, plot_scores, zdir='x', offset=0, cmap=plt.cm.coolwarm)
cset = ax.contour(X_, Y_, plot_scores, zdir='y', offset=1., cmap=plt.cm.coolwarm)
ax.set_zlim((-1, 1))
ax.set_xlabel('h')
ax.set_ylabel(r'$\lambda$')
ax.set_zlabel('Jensen divergence (lower is better)\nlog(J(hgl) / J(gl)')
ax.set_title('Jmin(hgl:alpha={}, h={}) = {}, Jmin(gl, alpha={}) = {}'.format(Y_.flat[np.argmin(scores.T)],
                                                                             X_.flat[np.argmin(scores.T)],
                                                                             np.min(scores), Y_.flat[np.argmin(score_gl.T)],
                                                                             np.min(score_gl)))
```
