```python
import itertools
import covariance_learn as covl
reload(covl)
import numpy as np
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import matplotlib.colors as colors
import matplotlib.markers as markers
```

```python
mpl.rc('lines', linewidth=2, markersize=12)
mpl.rc('figure', figsize=(20,12))
def mean_value_list(L):
    L_ = [np.array(a) for a in L]
    n = max([a.shape[0] for a in L_])
    return np.nanmean(np.vstack([np.concatenate((np.array(a), np.nan * np.ones((n + 1 - a.shape[0], ))))[:-1]
                                  if a.shape[0] < n else a for a in L_]), axis=0)
```

```python
alpha = np.logspace(-2., 0., 3.)
rho = np.logspace(-2., 2., 3)
adaptive = [True, False]
optimal = [True, False]
p = 100
n = 10
n_ = 100

f = list()
for _ in range(5):
    Y = np.random.normal(size=(n, p))
    alpha_max = covl.alpha_max(Y)
    for (alpha_, adap, opt) in itertools.product(alpha, adaptive, optimal):
        alpha__ = alpha_ / alpha_max
        if not opt:
            for rho_ in rho:
                if adap:
                    gl = covl.GraphLasso(alpha__, rho=rho_, mu=2.)
                elif not adap:
                    gl = covl.GraphLasso(alpha__, rho=rho_)
                f.append({'alpha': alpha_, 'rho': rho_, 'adaptive': adap, 'func_vals': gl.fit(Y).f_vals_})
        else:
            if adap:
                gl = covl.GraphLasso(alpha__, rho=alpha__, mu=2.)
            elif not adap:
                gl = covl.GraphLasso(alpha__, rho=alpha__)
            f.append({'alpha': alpha_, 'rho': 'optimal', 'adaptive': adap, 'func_vals': gl.fit(Y).f_vals_})
df = pd.DataFrame(f)
```

```python
leg1 = list()
leg2 = list()
alphas = np.unique(np.array(df['alpha']))
n_alphas = alphas.size
colors_ = colors.hsv_to_rgb(np.dstack((np.linspace(0., .5, n_alphas), .9 * np.ones(n_alphas), .8 * np.ones(n_alphas))))
colors_ = colors_.squeeze()
rhos = np.unique(np.array(df['rho']))
print rhos
n_rhos = rhos.size
for (alpha_, rho_, adaptive_), group in df.groupby(['alpha', 'rho', 'adaptive']):
    if adaptive_: 
        plt.subplot(121)
        leg1.append('rho = {}, alpha = {}'.format(rho_, alpha_))
    else:
        plt.subplot(122)
        leg2.append('rho = {}, alpha = {}'.format(rho_, alpha_))
    plt.semilogx(mean_value_list(list(group['func_vals'])),
               color=colors_[alphas==alpha_, :].squeeze(),
               marker=markers.MarkerStyle().filled_markers[np.where(rhos==rho_)[0] + 1])
plt.subplot(121)
plt.xlabel('iterations', fontsize=24)
plt.ylabel(r'$[f + g](x^{k+1})$', fontsize=24)
plt.title('adaptive')
plt.legend(leg1)
plt.xlim([0., 100])
plt.ylim([-500., 500])
plt.subplot(122)
plt.xlabel('iterations', fontsize=24)
plt.ylabel(r'$[f + g](x^{k+1})$', fontsize=24)
plt.title('fixed')
plt.legend(leg2)
plt.xlim([0., 100])
plt.ylim([-500., 500])
plt.show()
```
