```python
import covariance_learn
import itertools
from mpl_toolkits.mplot3d import Axes3D
import phase_transition
import htree
import matplotlib.pyplot as plt
import numpy as np
import scipy.linalg
import scipy.special
from functools import partial 
%matplotlib inline
# from IPython.html.widgets import interact
```

```python
def _alpha_func(alpha, lev, h=1., max_level=1.):
    C = scipy.special.binom(max_level - 1, max_level - lev)
    beta1 = scipy.special.beta(max_level - lev + h, lev)
    beta2 = scipy.special.beta(h, max_level)
    return alpha * C * beta1 / beta2

def alpha_func_(h, max_level):
    return partial(_alpha_func, h=h, max_level=max_level)
```

```python
z1 = 2
z2 = 2
tree = htree.construct_tree(arity=4, depth=2)
# Create sample

a = -.4
b = .28
Theta = phase_transition._get_mx(a, b, mx_type='smith')

n_samples = 16

X = np.random.normal(size=(n_samples, Theta.shape[0]))

eigvals, eigvecs = scipy.linalg.eigh(Theta)
print np.min(eigvals)
C = eigvecs.dot(np.diag(1 / np.sqrt(eigvals))).dot(eigvecs.T)

# X = X - X.mean(axis=0)
X = X.dot(C)

Y = C

def cov2corr(cov):
    return np.diag(1 / np.sqrt(np.diag(cov))).dot(cov.dot(np.diag(1 / np.sqrt(np.diag(cov)))))

plt.figure()
plt.matshow(Theta, cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('population precision matrix')
plt.matshow(np.linalg.inv(cov2corr(X.T.dot(X))), cmap=plt.cm.RdBu_r)
plt.clim((-2, 2))
plt.colorbar()
plt.title('sample (scaled) precision matrix')
plt.matshow(cov2corr(Y.T.dot(Y)), cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('population correlation matrix')
plt.matshow(cov2corr(X.T.dot(X)), cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('sample correlation matrix')
plt.show()
```

```python
plt.figure()
levels_ = np.arange(1, 50)
plt.plot(levels_, _alpha_func(alpha=1, lev=levels_, h=1, max_level=49))
```

```python
max_level = max([lev for (_, lev) in tree.root_.get_descendants()])
# print "tree with {} levels".format(max_level)

n_h = 11
n_a = 11
h_vals = np.linspace(1e-3, 1, n_h)
h_ix = np.arange(n_h)
alpha_vals = np.logspace(-2, 0, n_a)
a_ix = np.arange(n_a)
scores = {'KL': np.zeros((n_h, n_a)), 'ell0': np.zeros((n_h, n_a))}
score_gl = {'KL': np.zeros((n_h, n_a)), 'ell0': np.zeros((n_h, n_a))}

for (ix, vals) in zip(itertools.product(h_ix, a_ix), itertools.product(h_vals, alpha_vals)):
    # print "computing h={}, alpha={} yielding ({}, {})".format(vals[0], vals[1],
    #                                                          alpha_func_(vals[0], max_level)(vals[1], 2),
    #   
    for score_norm in {'KL', 'ell0'}:
        if ix[0] == 0:
            covl = covariance_learn.GraphLasso(
                alpha=vals[1], score_norm=score_norm, max_iter=1e4, rho=2, verbose=50)
            score_gl[score_norm][..., ix[1]] = covl.fit(X).score(Y)
        covl = covariance_learn.HierarchicalGraphLasso(
            alpha=vals[1], htree=tree, alpha_func=alpha_func_(vals[0], max_level),
            score_norm=score_norm, max_iter=1e4, rho=2, verbose=50)
        scores[score_norm][ix] = covl.fit(X).score(Y)
    
```

## Jensen's divergence
### Hierarchical graph lasso
Showing the results of a grid search over the space $(\lambda, h)$ to minimise the symmetrised Kullback-Leibler (Jensen's) divergence

```python
fig = plt.figure()
plot_scores = scores['KL'].T
ax = fig.gca(projection='3d')
X_, Y_ = np.meshgrid(h_vals, alpha_vals)
surf = ax.plot_surface(X_, Y_, plot_scores, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cset = ax.contour(X_, Y_, plot_scores, zdir='z', offset=0, cmap=plt.cm.coolwarm)
cset = ax.contour(X_, Y_, plot_scores, zdir='x', offset=0, cmap=plt.cm.coolwarm)
cset = ax.contour(X_, Y_, plot_scores, zdir='y', offset=1., cmap=plt.cm.coolwarm)
ax.set_zlim((0, 25))
ax.set_xlabel('h')
ax.set_ylabel(r'$\lambda$')
ax.set_zlabel("Jensen's divergence (lower is better)\nlog(J(hgl) / J(gl)")
ax.set_title('Jmin(hgl:lambda={}, h={}) = {}, Jmin(gl, lambda={}) = {}'.format(Y_.flat[np.argmin(scores['KL'].T)],
                                                                               X_.flat[np.argmin(scores['KL'].T)],
                                                                               np.min(scores['KL']),
                                                                               Y_.flat[np.argmin(score_gl['KL'].T)],
                                                                               np.min(score_gl['KL'])))
```

### Comparing Hierarchical to plain graphical lasso

```python
fig = plt.figure()
plot_scores = np.log(scores['KL'] / score_gl['KL']).T
ax = fig.gca(projection='3d')
X_, Y_ = np.meshgrid(h_vals, alpha_vals)
surf = ax.plot_surface(X_, Y_, plot_scores, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cset = ax.contour(X_, Y_, plot_scores, zdir='z', offset=-1, cmap=plt.cm.coolwarm)
cset = ax.contour(X_, Y_, plot_scores, zdir='x', offset=0, cmap=plt.cm.coolwarm)
cset = ax.contour(X_, Y_, plot_scores, zdir='y', offset=1., cmap=plt.cm.coolwarm)
ax.set_zlim((-1, 1))
ax.set_xlabel('h')
ax.set_ylabel(r'$\lambda$')
ax.set_zlabel('Jensen divergence (lower is better)\nlog(J(hgl) / J(gl)')
ax.set_title('Jmin(hgl:lambda={}, h={}) = {}, Jmin(gl, lambda={}) = {}'.format(Y_.flat[np.argmin(scores['KL'].T)],
                                                                               X_.flat[np.argmin(scores['KL'].T)],
                                                                               np.min(scores['KL']),
                                                                               Y_.flat[np.argmin(score_gl['KL'].T)],
                                                                               np.min(score_gl['KL'])))
alpha_star = Y_.flat[np.argmin(scores['KL'].T)]
h_star = X_.flat[np.argmin(scores['KL'].T)]
h_star_gl = X_.flat[np.argmin(score_gl['KL'].T)]
alpha_star_gl = Y_.flat[np.argmin(score_gl['KL'].T)]
```

## Estimator of the precision matrix
The estimator of $\Theta$ is based on a hierarchical graphical lasso with embedded hierarchies.
The penalty function is $$\lambda^{\frac{\text{lev}_{\text{max}} - 1 + 10^h}{\text{lev} - 1 + 10^h}}$$ where $h$ is a free parameter, $\lambda$ is the usual penalisation, and lev is the level (depth) in the tree.

```python
max_level = max([lev for (_, lev) in tree.root_.get_descendants()])
def estimate_precision(alpha, h=None, method='hgl'):
    alpha_func = partial(_alpha_func, h=h, max_level=max_level)
    if method == 'gl':
        covl = covariance_learn.GraphLasso(alpha=alpha, score_norm='KL', max_iter = 1e4, rho=2, verbose=50)
    elif method == 'hgl':
        covl = covariance_learn.HierarchicalGraphLasso(alpha=alpha, htree=tree, alpha_func=alpha_func, score_norm='KL',
                                                       max_iter = 1e4, rho=2, verbose=50)
    covl.fit(X)
    K = covl.precision_
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.matshow(np.ma.masked_equal(K * (covl.auxiliary_prec_ != 0), 0), cmap=plt.cm.RdBu_r, vmin=-np.max(np.abs(K)) / 5.,
                vmax=np.max(np.abs(K)) / 5.)
    # ax1.set_clim((-np.max(np.abs(K)), np.max(np.abs(K))))  # (-1, 1))  # 
    # ax1.colorbar()
    ax1.set_title('Estimated precision')
    ax1.set_axis_bgcolor('.7')
    # inspire colormap by http://scikit-learn.org/stable/auto_examples/covariance/plot_sparse_cov.html
    ax2.matshow(np.array(covl.auxiliary_prec_ != 0, dtype=float), cmap=plt.cm.gray, vmin=0, vmax=1)
    DiagK = np.diag(np.diag(K))
    ax2.set_title('Support (derived from split variable)')
    fig, ax = plt.subplots()
    plt.plot(covl.f_vals_, 'b-')
    plt.title("Jensen's divergence = {}".format(covl.score(Y)))
    ax.set_ylabel('energy', color='b')
    for tl in ax.get_yticklabels():
        tl.set_color('b')
    ax2 = ax.twinx()
    ax2.semilogy(covl.var_gap_, 'g-')
    ax2.semilogy(covl.dual_gap_, 'g--')
    ax2.set_ylabel('split diff & dual update norm', color='g')
    for tl in ax2.get_yticklabels():
        tl.set_color('g')

# interact(estimate_precision, alpha=(0, 1., .01), h=(-2, 2, 1. / 8), method=('hgl', 'gl'))
```

Hierarchical Graph Lasso with $\lambda=\lambda^{\star}_{hgl}$ and $h=h^{\star}_{hgl}$ (both with respect to Jensen's divergence), possibly resulting in a suppression of some off-diagonal blocks that are supposed non-active

```python
estimate_precision(alpha=alpha_star, h=h_star, method='hgl')
```

Hierarchical Graph Lasso with $\lambda=\lambda^{\star}_{gl}$ and $h=h^{\star}|_{\lambda^{\star}_{gl}}$ (both with respect to Jensen's divergence), possibly resulting in a suppression of some off-diagonal blocks that are supposed non-active

```python
estimate_precision(alpha=alpha_star_gl, h=h_star_gl, method='hgl')
```

Graph Lasso with $\lambda=\lambda^{\star}_{hgl}$ from the Hierarchical graph lasso (with respect to Jensen's divergence)

```python
estimate_precision(alpha=alpha_star, method='gl')
```

Graph Lasso with $\lambda=\lambda^{\star}_{gl}$ (with respect to Jensen's divergence)

```python
estimate_precision(alpha=alpha_star_gl, method='gl')
```
