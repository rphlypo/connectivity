## Estimator of the precision matrix
The estimator of $\Theta$ is based on a hierarchical graphical lasso with embedded hierarchies.
The penalty function is $$\lambda^{\frac{\text{lev}_{\text{max}} - 1 + 10^h}{\text{lev} - 1 + 10^h}}$$ where $h$ is a free parameter, $\lambda$ is the usual penalisation, and lev is the level (depth) in the tree.

```python
def estimate_precision(alpha, h, method):
    max_level = 2
    alpha_func = lambda a, lev: a ** ((max_level - 1 + 10 ** h) / (lev - 1 + 10 ** h))
    tree = [range(4 * k, 4 * (k + 1)) for k in range(4)]
    if method == 'gl':
        covl = covariance_learn.GraphLasso(alpha=alpha)
    elif method == 'hgl':
        covl = covariance_learn.HierarchicalGraphLasso(alpha=alpha, htree=tree, alpha_func=alpha_func)
    covl.fit(X)
    K = covl.precision_
    plt.matshow(K, cmap=plt.cm.RdBu_r)
    plt.clim((-np.max(np.abs(K)) / 4., np.max(np.abs(K)) / 4.))  # (-.1, .1))  # 
    plt.colorbar()
    
    fig, ax = plt.subplots()
    plt.plot(covl.f_vals_, 'b-')
    ax.set_ylabel('energy', color='b')
    for tl in ax.get_yticklabels():
        tl.set_color('b')
    ax2 = ax.twinx()
    ax2.semilogy(covl.var_gap_, 'g-')
    ax2.semilogy(covl.dual_gap_, 'g--')
    ax2.set_ylabel('split diff & dual update norm', color='g')
    for tl in ax2.get_yticklabels():
        tl.set_color('g')

# interact(estimate_precision, alpha=(0, 1., .01), h=(-2, 2, 1. / 8), method=('hgl', 'gl'))
```

Hierarchical Graph Lasso with $\lambda=1/10$ and $h=8/10$, resulting in a suppression of the off-diagonal blocks that are supposed non-active

```python
estimate_precision(alpha=.1, h=.8, method='hgl')
```

Hierarchical Graph Lasso with $\lambda=1/10$ and $h=-100$, which approaches the Graphical Lasso for a tree with depth two.

```python
estimate_precision(alpha=.1, h=-100, method='hgl')
```

```python
estimate_precision(alpha=.1, h=-2, method='gl')
```
