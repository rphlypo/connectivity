```python
# Initiate precision matrix
Z1 = np.random.normal(size=(6, 2))
Z2 = np.random.normal(size=(9, 3))
Z = scipy.linalg.block_diag(Z1.T.dot(Z1) / Z1.shape[1], Z2.T.dot(Z2) / Z2.shape[1])
# covariance matrix
Cov = scipy.linalg.inv(Z)
# Create sample
eigvals, eigvecs = scipy.linalg.eigh(Cov)
M = eigvecs.dot(np.diag(np.sqrt(eigvals))).dot(eigvecs.T)
X = np.random.normal(size=(15, 5))
X = X.dot(M.T)

# dependence tree
tree_list = [[k for k in np.arange(Z1.shape[1])], [k + Z1.shape[1] for k in np.arange(Z2.shape[1])]]
def plot(alpha):
    hgl = covariance_learn.HierarchicalGraphLasso(tree_list, alpha)
    hgl.fit(X)
    plt.matshow(hgl.precision_)
    print hgl.score(X)
    plt.title("alpha = {}".format(alpha))
    plt.colorbar()
    plt.clim((-2., 2.))
    plt.figure()
    plt.plot(hgl.f_vals_)
    plt.title("penalised log-likelihood (wrong penalisation parameters!!)")
    plt.show()

interact(plot, alpha=(0, 1.0, .01))
```
