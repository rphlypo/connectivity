```python
z1 = 2
z2 = 2
tree = [[k for k in np.arange(z1)], [k + z1 for k in np.arange(z2)]]
# Create sample

a = .3
b = .3
Theta = phase_transition._get_mx(a, b, mx_type='smith')

X = np.random.normal(size=(30, Theta.shape[0]))

eigvals, eigvecs = scipy.linalg.eigh(Theta)
C = eigvecs.dot(np.diag(1 / np.sqrt(eigvals))).dot(eigvecs.T)

# X = X - X.mean(axis=0)
X = X.dot(C)

Y = C

plt.matshow(np.linalg.inv(X.T.dot(X)))
```

## Estimator of the precision matrix
The estimator of $\Theta$ is based on a hierarchical graphical lasso with embedded hierarchies.

```python
def estimate_precision(alpha, h, method):
    alpha_func = lambda a, lev: a ** (h * (2 + 1. / h - lev))
    tree = [range(4 * k, 4 * (k + 1)) for k in range(4)]
    if method == 'gl':
        covl = covariance_learn.GraphLasso(alpha=alpha)
    elif method == 'hgl':
        covl = covariance_learn.HierarchicalGraphLasso(alpha=alpha, htree=tree, alpha_func=alpha_func)
    covl.fit(X)
    K = covl.precision_
    plt.matshow(K, cmap=plt.cm.RdBu_r)
    plt.clim((-np.max(np.abs(K)) / 4., np.max(np.abs(K)) / 4.))
    plt.colorbar()
    
    fig, ax = plt.subplots()
    plt.plot(covl.f_vals_, 'b-')
    ax.set_ylabel('energy', color='b')
    for tl in ax.get_yticklabels():
        tl.set_color('b')
    ax2 = ax.twinx()
    ax2.semilogy(covl.var_gap_, 'g-')
    ax2.semilogy(covl.dual_gap_, 'g--')
    ax2.set_ylabel('split diff & dual update norm', color='g')
    for tl in ax2.get_yticklabels():
        tl.set_color('g')

interact(estimate_precision, alpha=(0, 1., .01), h=(1, 20, 1), method=('hgl', 'gl'))
# estimate_precision(alpha=.3, h=10, method='hgl')
```

```python
alpha_star_hgl, LL_hgl = covariance_learn.cross_val(X, method='hgl', htree=tree, alpha_tol=1e-3, n_iter=10, train_size=.2,
                                                     verbose=30)
score_hgl = covariance_learn.HierarchicalGraphLasso(tree, alpha=alpha_star_hgl).fit(Y).score(Y)
alpha_star_gl, LL_gl = covariance_learn.cross_val(X, method='gl', alpha_tol=1e-3, n_iter=10, train_size=.2, verbose=30)
score_gl = covariance_learn.GraphLasso(alpha=alpha_star_gl).fit(Y).score(Y)

print 'scores:\n\thgl = {}\n\t gl = {}'.format(score_hgl, score_gl)

plt.plot(np.array([LL_hgl, LL_gl]).T)
```

```python
z1 = 2
z2 = 2
# Create sample

a = (3. - np.sqrt(5)) / 2. + .01
b = a - .02
M = scipy.linalg.inv( np.array([[1., a, b, 0.], [a, 1., 0., 0.], [b, 0., 1., a], [0., 0., a, 1.]]))

eigvals, eigvecs = scipy.linalg.eigh(M)
M = eigvecs.dot(np.diag(np.sqrt(eigvals))).dot(eigvecs.T)

X = np.random.normal(size=(50, 4))
# X = X - X.mean(axis=0)
X = X.dot(M)

# Y = np.random.normal(size=(10, 4))
# Y = Y - Y.mean(axis=0)
Y = M

plt.matshow(scipy.linalg.inv(X.T.dot(X) / X.shape[0]))
plt.colorbar()

# dependence tree
tree_list = [[k for k in np.arange(z1)], [k + z1 for k in np.arange(z2)]]

def plot(alpha, method='hgl'):
    if method == 'hgl':
        hgl = covariance_learn.HierarchicalGraphLasso(tree_list, alpha, rho=1., tol=1e-12, mu=2.)
    elif method == 'gl':
        hgl = covariance_learn.GraphLasso(alpha, rho=1., tol=1e-12, mu=2.)
    hgl.fit(X)
    plt.matshow(hgl.precision_)
    plt.title("alpha = {}, LL = {}".format(alpha, hgl.score(Y)))
    plt.colorbar()
    cmax = np.max(np.abs(hgl.precision_))
    plt.clim((-cmax, cmax))
    fig, ax = plt.subplots()
    plt.plot(hgl.f_vals_, 'b-')
    ax.set_ylabel('energy', color='b')
    for tl in ax.get_yticklabels():
        tl.set_color('b')
    ax2 = ax.twinx()
    ax2.semilogy(hgl.var_gap_, 'g-')
    ax2.semilogy(hgl.dual_gap_, 'g--')
    ax2.set_ylabel('split diff & dual update norm', color='g')
    for tl in ax2.get_yticklabels():
        tl.set_color('g')

    plt.title("penalised log-likelihood")
    plt.show()

interact(plot, alpha=(0, 1., .01), method='hgl')
interact(plot, alpha=(0, 1., .01), method='gl')
```
