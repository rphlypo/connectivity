```python
import covariance_learn
import phase_transition
import htree
import matplotlib.pyplot as plt
import numpy as np
import scipy.linalg
import compute_hierarchical_parcellation as chp
from functools import partial
%matplotlib inline
# from IPython.html.widgets import interact
```

```python
max_level = 4
tree = htree.construct_tree(arity=2, depth=max_level)
# Create sample

a = -.4
b = .3
Theta = phase_transition._get_mx(a, b, mx_type='smith')

n_samples = 16

X = np.random.normal(size=(n_samples, Theta.shape[0]))

eigvals, eigvecs = scipy.linalg.eigh(Theta)
print np.min(eigvals)
C = eigvecs.dot(np.diag(1 / np.sqrt(eigvals))).dot(eigvecs.T)

# X = X - X.mean(axis=0)
X = X.dot(C)

Y = C

def cov2corr(cov):
    return np.diag(1 / np.sqrt(np.diag(cov))).dot(cov.dot(np.diag(1 / np.sqrt(np.diag(cov)))))

plt.matshow(Theta, cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('population precision matrix')
plt.matshow(np.linalg.inv(cov2corr(X.T.dot(X))), cmap=plt.cm.RdBu_r)
plt.clim((-2, 2))
plt.colorbar()
plt.title('sample (scaled) precision matrix')
plt.matshow(cov2corr(Y.T.dot(Y)), cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('population correlation matrix')
plt.matshow(cov2corr(X.T.dot(X)), cmap=plt.cm.RdBu_r)
plt.clim((-1, 1))
plt.colorbar()
plt.title('sample correlation matrix')
```

## Estimator of the precision matrix
The estimator of $\Theta$ is based on a hierarchical graphical lasso with embedded hierarchies.
The penalty function is $$\lambda^{\frac{\text{lev}_{\text{max}} - 1 + 10^h}{\text{lev} - 1 + 10^h}}$$ where $h$ is a free parameter, $\lambda$ is the usual penalisation, and lev is the level (depth) in the tree.

```python
def _alpha_func(a, lev, h=None, max_level=None):
    C = scipy.special.binom(max_level, lev)
    beta1 = scipy.special.beta(lev + h, max_level - lev + 1)
    beta2 = scipy.special.beta(h, 1)
    return a * C * beta1 / beta2

def estimate_precision(alpha, h, method):
    alpha_func = partial(_alpha_func, h=h, max_level=max_level)
    if method == 'gl':
        covl = covariance_learn.GraphLasso(alpha=alpha, mu=2, score_norm='KL', max_iter = 1e3)
    elif method == 'hgl':
        covl = covariance_learn.HierarchicalGraphLasso(alpha=alpha, htree=tree, alpha_func=alpha_func, mu=2, score_norm='KL',
                                                       max_iter = 1e3)
    covl.fit(X)
    K = covl.precision_
    plt.matshow(K, cmap=plt.cm.RdBu_r)
    DiagK = np.diag(np.diag(K))
    plt.clim((-np.max(np.abs(K)), np.max(np.abs(K))))  # (-1, 1))  # 
    plt.colorbar()
    
    fig, ax = plt.subplots()
    plt.plot(covl.f_vals_, 'b-')
    plt.title("Jensen's divergence = {}".format(covl.score(Y)))
    ax.set_ylabel('energy', color='b')
    for tl in ax.get_yticklabels():
        tl.set_color('b')
    ax2 = ax.twinx()
    ax2.semilogy(covl.var_gap_, 'g-')
    ax2.semilogy(covl.dual_gap_, 'g--')
    ax2.set_ylabel('split diff & dual update norm', color='g')
    for tl in ax2.get_yticklabels():
        tl.set_color('g')

# interact(estimate_precision, alpha=(0, 1., .01), h=(-2, 2, 1. / 8), method=('hgl', 'gl'))
```

Hierarchical Graph Lasso with $\lambda=1/1000$ and $h=6/10$, resulting in a suppression of some off-diagonal blocks that are supposed non-active

```python
estimate_precision(alpha=.001, h=.7, method='hgl')
```

Hierarchical Graph Lasso with $\lambda=1/1000$ and $h=-100$, which approaches the Graphical Lasso for a tree with depth two.

```python
estimate_precision(alpha=.001, h=, method='hgl')
```

Indeed, the Graphical Lasso with the same $\lambda$ as in the above example of Hierarchical Graph Lasso yields the same results when the latter has its parameter $h\to -\infty$

```python
estimate_precision(alpha=.001, h=-2, method='gl')
```

To suppress the blocks in the Graphical Lasso we need to choose a higher $\lambda$ penalising the diagonal blocks more. It is not sure that a such $\lambda\neq\lambda_{\text{max}}$ actually exists.

```python
estimate_precision(alpha=.001, h=-10, method='gl')
```
